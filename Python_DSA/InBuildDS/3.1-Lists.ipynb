```python
### Accessing List Elements

fruits=["apple","banana","cherry","kiwi","gauva"]
print(len(fruits))
```

```python
print(fruits[0])
print(fruits[2])
print(fruits[4])
print(fruits[-1]) #Accessing last element
```

```python
print(fruits[1:]) #from index 1 to end
print(fruits[1:3]) #from index 1 to index 2
print(fruits[:3]) #from start to index 2
print(fruits[:]) #from start to end 
print(fruits[-4:-1]) #negative indexing
print(fruits[::-1]) #reversing the list
print(fruits[::2]) #accessing alternate elements
print(fruits[1::2]) #accessing alternate elements starting from index 1
print(fruits[-1::-1]) #reversing the list using negative indexing
print(fruits[-2::-2]) #accessing alternate elements in reverse order starting from second last element
print(fruits[-1:-4:-1]) #accessing last three elements in reverse order
print(fruits[-3:]) #accessing last three elements
print(fruits[:-4]) #accessing all elements except last three elements
print("---------------------------------------------------------")
print(fruits[-len(fruits):]) #accessing all elements using length of the list
print(fruits[len(fruits)-3:]) #accessing last three elements using length of the list
print(fruits[:len(fruits)-2]) #accessing all elements except last two elements using length of the list
print(fruits[::len(fruits)-1]) #accessing first and last element using length of the list
print(fruits[-len(fruits)::len(fruits)-1]) #accessing first and last element using negative indexing and length of the list
print(fruits[len(fruits)-1::-len(fruits)+1]) #accessing last and first element using length of the list in reverse order
print(fruits[-1:len(fruits)-len(fruits)-1:-1]) #accessing last to first element in reverse order using negative indexing and length of the list
print(fruits[1:len(fruits)-1]) #accessing all elements except first and last element using length of the list
print(fruits[-len(fruits)+1:-1]) #accessing all elements except first and last element using negative indexing and length of the list
print(fruits[::len(fruits)//2]) #accessing first, middle and last element using length of the list
print(fruits[-len(fruits)::len(fruits)//2]) #accessing first,middle and last element using negative indexing and length of the list
print(fruits[len(fruits)//2::-len(fruits)//2]) #accessing middle, first and last element using length of the list in reverse order
print(fruits[-(len(fruits)//2):len(fruits)-(len(fruits)//2)+1]) #accessing middle, last and first element using negative indexing and length of thelist
print(fruits[1:len(fruits):2]) #accessing alternate elements starting from index 1 using length of the list 
print(fruits[-len(fruits)+1::2]) #accessing alternate elements starting from index 1 using negative indexing and length of the list
print(fruits[len(fruits)-2::-2]) #accessing alternate elements in reverse order
print(fruits[-2:len(fruits)-len(fruits)-1:-2]) #accessing alternate elements in reverse order using negative indexing and length of the list
print("---------------------------------------------------------")
print(fruits[::3]) #accessing every third element
print(fruits[-len(fruits)::3]) #accessing every third element using negative indexing and length of the list
print(fruits[3::-3]) #accessing every third element in reverse order
print(fruits[-1:len(fruits)-len(fruits)-1:-3]) #accessing every third element in reverse order using negative indexing and length of the list
print(fruits[1:len(fruits):3]) #accessing every third element starting from index 1 using length of the list
print(fruits[-len(fruits)+1::3]) #accessing every third element starting from index 1 using negative indexing and length of the list
print(fruits[2:len(fruits):3]) #accessing every third element starting from index 2 using length of the list
print(fruits[-len(fruits)+2::3]) #accessing every third element starting from index 2 using negative indexing and length of the list
print(fruits[2::-3]) #accessing every third element in reverse order starting from index 2
print(fruits[-2:len(fruits)-len(fruits)-1:-3]) #accessing every third element in reverse order starting from index 2 using negative indexing and length of the list     

```

##### List Comprehension

Basics Syantax            [expression for item in iterable]

with conditional logic    [expression for item in iterable if condition]

Nested List Comprehension [expression for item1 in iterable1 for item2 in iterable2]




```python
### Basic List Comphrension

sqaure=[num**2 for num in range(10)] #combine the for loop and the squaring operation into a single line
print(sqaure)
```

```python
### List Comprehension with Condition
lst=[]
for i in range(10):
    if i%2==0:
        lst.append(i)

print(lst)
        
```

```python
even_numbers=[num for num in range(10) if num%2==0] #Using list comprehension to create a list of even numbers from 0 to 9
print(even_numbers)
```

```python
## Nested List Comphrension

lst1=[1,2,3,4]
lst2=['a','b','c','d']

pair=[[i,j] for i in lst1 for j in lst2]

print(pair)
```

```python
## List Comprehension with function calls
words = ["hello", "world", "python", "list", "comprehension"]
lengths = [len(word) for word in words]
print(lengths)  # Output: [5, 5, 6, 4, 13]

```

```python
#List comprehension with if else condition
values = [10, 15, 20, 25, 30]
result = [val * 2 if val > 20 else val / 2 for val in values]
print(result)  # Output: [5.0, 7.5, 20, 50, 60]
```

#### Conclusion
List comprehensions are a powerful and concise way to create lists in Python. They are syntactically compact and can replace more verbose looping constructs. Understanding the syntax of list comprehensions will help you write cleaner and more efficient Python code.

```python
import numpy as np
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

transposed = []

# We use len(matrix[0]) to determine the number of columns
# This makes the code work even if the matrix isn't 3x3
for i in range(len(matrix[0])):
    print(f"We are going to take the {i}-th element of each row")
    lst = []
    
    for row in matrix:
        print(f"  The row we are considering: {row} and we are going to add {row[i]} to a temp list")
        # FIX: This line must be indented to be INSIDE the inner loop
        lst.append(row[i])
        
    print(f"  The {i}-th column elements (new row) are: {lst}")
    
    # FIX: This must be indented to be INSIDE the outer loop, 
    # but OUTSIDE the inner loop
    transposed.append(lst)

print("=" * 50)
print("Transposed Matrix:", transposed)
# NumPy arrays print as beautiful matrices by default
np_matrix = np.array(transposed)
print(np_matrix)

```

```python
matrix = [
    [1, 2, 3],          
    [4, 5, 6],
    [7, 8, 9]  
] 

transposed = []
for i in range(len(matrix)):      # The Outer Loop the outer loop range(len(matrix)) iterates based on the row count of the original matrix. To transpose correctly, you must iterate based on the column count (the length of the first row).
    
    new_row = []
    for row in matrix:                # The Inner Loop
        new_row.append(row[i])        # Grabbing the i-th element
    transposed.append(new_row)  
print(transposed)
```

```python
#Transposing a matrix using list comprehension
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
transposed = [[row[i] for row in matrix] for i in range(len(matrix))]
print(transposed)
```

```python
# We use len(matrix[0]) to iterate over the number of columns
transposed = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
print(transposed)
```
