```python
### Variable Length Arguments
## Positional And Keywords arguments

def print_numbers(*krish):
    for number in krish:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Krish")
```

```python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Krish")
```

```python
### Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Krish",age="32",country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional arument :{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"Krish",name="Krish",age="32",country="India")
```

```python
# Python program to print Highest Common Factor (HCF) of two numbers

def computeHCF(a, b):
    smaller =b if a>b else a

    hcf=1
    for i in range(1,smaller+1): 
        if (a%i ==0) and (b%i==0):
            hcf =i
    return hcf

num1=12
num2=14
    
print ("H.C.F of {0} and {1} is : {2}" .format(num1, num2, computeHCF (num1, num2)))


```

```python
import math

num1 = 12
num2 = 14

# Using the built-in function
result = math.gcd(num1, num2)

print(f"H.C.F of {num1} and {num2} is : {result}")
```

```python
# The enumerate() function is a powerful tool that adds a counter to an iterable. It returns an enumerate object, which you can use in a for loop to get both the index and the value of each item.
tasks = ['Review code', 'Write documentation', 'Deploy to production']

# Using enumerate to get both the index and the task
print("Today's Tasks:")
for index, task in enumerate(tasks):
    print(f"{index}: {task}")
```

```python
tasks = ['Review code', 'Write documentation', 'Deploy to production']

# Using enumerate with a starting count of 1
print("Today's Tasks (Numbered):")
for count, task in enumerate(tasks, start=1): # If you want your list to start from 1 instead of 0, you can use the optional start argument.
    print(f"{count}. {task}")
```

```python
# Before enumerate, you would have to create and manage a counter variable
tasks = ['Review code', 'Write documentation', 'Deploy to production']
index = 1
for task in tasks:
    print(f"{index}: {task}")
    index += 1 # It's easy to forget this line!
```

The reduce() function applies a given function of two arguments cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value

```python
from functools import reduce 

def multiply(x,y):
    return x*y;

product = reduce(multiply, lst)

print(product)
```

```python
numbers = [1, 2, 3, 4, 5]

# Good, readable, and fast
total = sum(numbers)
print(total) # Output: 15

# Bad, less readable
total = reduce(lambda x, y: x + y, numbers)
```

```python
from functools import reduce

set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7}
set3 = {3, 4, 5, 8, 9}
list_of_sets = [set1, set2, set3]

# Use reduce to find the intersection of all sets in the list
common_elements = reduce(lambda s1, s2: s1.intersection(s2), list_of_sets)

print(f"The common elements are: {common_elements}")
```

```python
#Python program to make a simple calculator that can add ,subtract,multiply and division

def add (a, b):
    return a+b 
def multiply (a, b) :
    return a*b
def subtract (a,b):
    return a-b
def division (a, b):
    return a/b

print ("Select Option")
print ("1. Addition")
print ("2. Subtraction")
print ("3. Multiplication")
print ("4. Division")

choice = int (input ("Enter Choice 1/2/3/4"))
print(f"You selected choice: {choice}")

num1 = float (input ("Enter first number :"))
num2 = float (input ("Enter Second number : "))
print(f"first number: {num1}")
print(f"Second number: {num2}")

if choice ==1 :
    print ("Addition of {0} and {1} is {2}".format (num1 ,num2 , add(num1 ,num2)))

elif choice ==2 :
    print ("Subtraction of {0} and {1} is {2}".format (num1 ,num2 , subtract(num1 ,num2)))

elif choice ==3 :
    print ("Multiplication of {0} and {1} is {2}".format (num1 ,num2 , multiply(num1 ,num2)))

elif choice ==4 :
    print ("Division of {0} and {1} is {2}".format (num1 ,num2 , division(num1 ,num2)))

else :

    print ("Invalid Choice")
```
